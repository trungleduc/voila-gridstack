```python
import pandas as pd
import numpy as np
import warnings

from scipy import stats
import matplotlib as mpl
import matplotlib.pyplot as plt
import seaborn as sns

%matplotlib inline
warnings.filterwarnings("ignore")
pd.set_option('display.max_rows', None)
sns.set(style="ticks", color_codes=True)

df = pd.read_csv('Iris.csv')
```

## Feature exploration
First of all, we have to know exactly which variables have our dataset, what they mean and how they influence the decision.
Furthermore, we have to see if there are missing values, not numeric variables and how this variable can be more effective.
So, we have to pry all the information that we can from every feature.

```python
df.info()
```

```python
pd.isnull(df).sum().sort_values(ascending=False)
```

```python
df.describe(include = 'all')
```

At this point we can use seaborm's FacetGrid to plot by species

```python
sns.FacetGrid(df, hue="Species").map(plt.scatter, "SepalLengthCm", "SepalWidthCm").add_legend();
```

Another useful plot is the pairplot, which shows the bivariate relation between each pair of features

```python
sns.pairplot(df.drop("Id", axis=1), hue="Species", size=3);
```
